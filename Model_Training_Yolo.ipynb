```python
import os
from ultralytics import YOLO

# ---------------------------------
# 1. –ó–ê–í–ê–ù–¢–ê–ñ–ï–ù–ù–Ø –ú–û–î–ï–õ–Ü
# ---------------------------------

# –ó–∞–≤–∞–Ω—Ç–∞–∂—É—î–º–æ –ø–æ–ø–µ—Ä–µ–¥–Ω—å–æ –Ω–∞–≤—á–µ–Ω—É –º–æ–¥–µ–ª—å.
# 'yolov8n.pt' - 'n' –æ–∑–Ω–∞—á–∞—î 'nano', –Ω–∞–π–º–µ–Ω—à–∞ —ñ –Ω–∞–π—à–≤–∏–¥—à–∞ –º–æ–¥–µ–ª—å.
# –í–æ–Ω–∞ –≤–∂–µ –≤–º—ñ—î –∑–Ω–∞—Ö–æ–¥–∏—Ç–∏ "–ª—é–¥–µ–π", "–º–∞—à–∏–Ω–∏" —Ç–æ—â–æ, –º–∏ —ó—ó –ø–µ—Ä–µ–≤—á–∏–º–æ.
model = YOLO('yolov8n.pt')

# ---------------------------------
# 2. –ù–ê–í–ß–ê–ù–ù–Ø –ú–û–î–ï–õ–Ü
# ---------------------------------

print("–ü–æ—á–∏–Ω–∞—é –Ω–∞–≤—á–∞–Ω–Ω—è YOLOv8...")

# –ó–∞–ø—É—Å–∫–∞—î–º–æ –Ω–∞–≤—á–∞–Ω–Ω—è
# data: —à–ª—è—Ö –¥–æ –Ω–∞—à–æ–≥–æ .yaml —Ñ–∞–π–ª—É
# epochs: –∫—ñ–ª—å–∫—ñ—Å—Ç—å –µ–ø–æ—Ö (–ø–æ—á–Ω–µ–º–æ –∑ 50, YOLO –≤—á–∏—Ç—å—Å—è —à–≤–∏–¥–∫–æ)
# imgsz: —Ä–æ–∑–º—ñ—Ä –∑–æ–±—Ä–∞–∂–µ–Ω–Ω—è (—Ç–æ–π —Å–∞–º–∏–π, —â–æ –º–∏ –≥–µ–Ω–µ—Ä—É–≤–∞–ª–∏)
# device: 0 - –∞–≤—Ç–æ–º–∞—Ç–∏—á–Ω–æ –≤–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É–≤–∞—Ç–∏ GPU
results = model.train(
    data=os.path.abspath('learning/yolo.yaml'),
    epochs=50,
    imgsz=640,
    device=0  # –í–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É–≤–∞—Ç–∏ GPU
)

print("--- –ù–∞–≤—á–∞–Ω–Ω—è YOLO –∑–∞–≤–µ—Ä—à–µ–Ω–æ! ---")
print(f"–†–µ–∑—É–ª—å—Ç–∞—Ç–∏ –∑–±–µ—Ä–µ–∂–µ–Ω–æ –≤ –ø–∞–ø—Ü—ñ: {results.save_dir}")
```
