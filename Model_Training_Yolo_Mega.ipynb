```python
import os
import yaml
from ultralytics import YOLO

# ---------------------------------
# 1. –ö–û–ù–§–Ü–ì–£–†–ê–¶–Ü–Ø
# ---------------------------------

DATA_DIR = os.path.abspath('learning/data_yolo_mega')
YAML_FILE_PATH = os.path.join('learning', 'yolo_mega.yaml') # –§–∞–π–ª –±—É–¥–µ –ø–µ—Ä–µ–∑–∞–ø–∏—Å–∞–Ω–æ

# !!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
# !!! –û–°–¨ –ì–û–õ–û–í–ù–ï –í–ò–ü–†–ê–í–õ–ï–ù–ù–Ø !!!
# !!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
# –¢–µ–ø–µ—Ä –º–∏ –≤–∫–∞–∑—É—î–º–æ –ª–∏—à–µ –û–î–ò–ù –∫–ª–∞—Å
CLASS_NAMES = ['digit']

print(f"–®–ª—è—Ö –¥–æ –¥–∞–Ω–∏—Ö: {DATA_DIR}")
print(f"–§–∞–π–ª –∫–æ–Ω—Ñ—ñ–≥—É—Ä–∞—Ü—ñ—ó –±—É–¥–µ —Å—Ç–≤–æ—Ä–µ–Ω–æ —Ç—É—Ç: {YAML_FILE_PATH}")
print(f"–ù–∞–≤—á–∞—î–º–æ –Ω–∞ –∫–ª–∞—Å–∏: {CLASS_NAMES}")

# ---------------------------------
# 2. –°–¢–í–û–†–ï–ù–ù–Ø .YAML –§–ê–ô–õ–£ (—Ç–µ–ø–µ—Ä –ø—Ä–∞–≤–∏–ª—å–Ω–æ–≥–æ)
# ---------------------------------

data_yaml = {
    'path': DATA_DIR,
    'train': 'train/images',
    'val': 'valid/images',
    'names': {i: name for i, name in enumerate(CLASS_NAMES)} # –¶–µ —Å—Ç–≤–æ—Ä–∏—Ç—å {0: 'digit'}
}

try:
    with open(YAML_FILE_PATH, 'w') as f:
        yaml.dump(data_yaml, f, sort_keys=False)
    print("–§–∞–π–ª yolo_mega.yaml —É—Å–ø—ñ—à–Ω–æ —Å—Ç–≤–æ—Ä–µ–Ω–æ (–¥–ª—è 1 –∫–ª–∞—Å—É).")
except Exception as e:
    print(f"–ü–æ–º–∏–ª–∫–∞ –ø—Ä–∏ —Å—Ç–≤–æ—Ä–µ–Ω–Ω—ñ .yaml —Ñ–∞–π–ª—É: {e}")

# ---------------------------------
# 3. –ù–ê–í–ß–ê–ù–ù–Ø –ú–û–î–ï–õ–Ü
# ---------------------------------

model = YOLO('yolov8n.pt')

print("--- –ü–æ—á–∞—Ç–æ–∫ –Ω–∞–≤—á–∞–Ω–Ω—è YOLOv8 (1 –ö–õ–ê–°) ---")

try:
    results = model.train(
        data=YAML_FILE_PATH,
        epochs=100, # –¢–µ–ø–µ—Ä —Ü–µ –±—É–¥–µ –Ω–∞–±–∞–≥–∞—Ç–æ —à–≤–∏–¥—à–µ
        imgsz=640,
        batch=16,
        name='yolo_single_class_run' # –ù–æ–≤–∞ –ø–∞–ø–∫–∞ –¥–ª—è –Ω–æ–≤–æ—ó –º–æ–¥–µ–ª—ñ
    )

    print("--- –ù–∞–≤—á–∞–Ω–Ω—è –∑–∞–≤–µ—Ä—à–µ–Ω–æ ---")

    # ---------------------------------
    # 4. –í–ê–õ–Ü–î–ê–¶–Ü–Ø –¢–ê –ó–ë–ï–†–ï–ñ–ï–ù–ù–Ø
    # ---------------------------------

    print("--- –ó–∞–ø—É—Å–∫ –≤–∞–ª—ñ–¥–∞—Ü—ñ—ó –Ω–∞ 'valid' —Å–µ—Ç—ñ ---")
    metrics = model.val()
    print("--- –í–∞–ª—ñ–¥–∞—Ü—ñ—é –∑–∞–≤–µ—Ä—à–µ–Ω–æ ---")

    print(f"–ù–∞–π–∫—Ä–∞—â—É –º–æ–¥–µ–ª—å –∑–±–µ—Ä–µ–∂–µ–Ω–æ –≤: runs/detect/yolo_single_class_run/weights/best.pt")

    print("--- –ï–∫—Å–ø–æ—Ä—Ç –º–æ–¥–µ–ª—ñ –≤ ONNX ---")
    model.export(format='onnx')
    print("--- –ú–æ–¥–µ–ª—å –µ–∫—Å–ø–æ—Ä—Ç–æ–≤–∞–Ω–æ ---")

except Exception as e:
    print(f"\n--- ‚ùå –ü–û–ú–ò–õ–ö–ê –ü–Ü–î –ß–ê–° –ù–ê–í–ß–ê–ù–ù–Ø ---")
    print(e)
```
